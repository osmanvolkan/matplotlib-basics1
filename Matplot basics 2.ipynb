# Stacked Vertical Bar

```python
x1=["Isaac","Asimov","John","Tolkien"]
y1=[17,18,29,40]
y2=[20,21,22,23]
plt.figure(figsize=(5,7))
plt.bar(x1,y1,label="Open Tickets",width=0.5,color="#FF6F00")
plt.bar(x1,y2,label="Closed Tickets",width=0.5,color="#FFB300")
plt.xlabel("X")
plt.ylabel("Y",rotation=0)
plt.title("Bar Chart")
plt.legend()
plt.show()
```

```python
x1=["Isaac","Asimov","John","Tolkien"]
y1=np.array([17,18,29,40])
y2=np.array([20,21,22,23])#if we dont insert numbers as arrays,we cannot
y3=np.array([5,9,11,12])  #place labels on top of each other using bottom command. 
plt.figure(figsize=(5,7))
plt.bar(x1,y1,label="Open Tickets",width=0.5,color="#FF6F00")
plt.bar(x1,y2,label="Closed Tickets",width=0.5,color="#FFB300",bottom=y1)
plt.bar(x1,y3,label="Cancelled",width=0.5,color="#F7DC6F",bottom=y1+y2)
plt.xlabel("X")
plt.ylabel("Y",rotation=0)
plt.title("Bar Chart")
plt.legend()
plt.show()
```

# Stacked Vertical Bar

```python
#Grouped bar chart
plt.figure(figsize=(8,5))
barWidth=0.25
y1=np.array([17,18,29,40])
y2=np.array([20,21,22,21])
plt.barh(x1,y1,label="Open Tickets",color="#FF6F00")
plt.barh(x1,y2,label="Closed Tickets",left=y1, color="#FFB300")

plt.xlabel("X")
plt.ylabel("Y",rotation=0)
plt.title("Bar Chart")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(7,9))
barWidth=0.25
y1=np.array([17,18,29,40])
y2=np.array([20,21,22,21])
plt.barh(x1,y1,label="Open Tickets",color="#FF6F00")
plt.barh(x1,y2,label="Closed Tickets",left=y1, color="#FFB300")

plt.xlabel("X")
plt.ylabel("Y",rotation=45)

for x,y in zip(x1,y1):
    plt.text(y-10,x,"%d"%y,ha="center",va="bottom")
for x,y,z in zip(x1,y2,y1):
    plt.text(y+z-10,x,"%d"%y,ha="center",va="bottom")    

plt.title("Bar Chart")
plt.legend()
plt.show()
```

# Scatter Graphs

```python
x1=np.array([250,150,350,252,450,550,455,358,158,355])
y1=np.array([50,50,80,90,100,50,60,88,45,54])
x2=np.array([200,100,300,220,400,500,450,380,160,350])
y2=np.array([400,500,800,900,1000,100,600,808,504,405])
```

```python
#graph-1
plt.scatter(x1,y1)
plt.xlabel("Time Spent",fontsize=12)
plt.ylabel("Score",fontsize=12)
plt.title("Scatter Graph")
plt.show()
```

```python
#Graph-2
plt.scatter(x1,y1,label="Class 1")
plt.scatter(x2,y2,label="Class 2",color="r")
plt.xlabel("Time Spent",fontsize=12)
plt.ylabel("Score",fontsize=12)
plt.title("Scatter Graph")
plt.legend()
plt.show()


```

```python
#Graph 3 
plt.scatter(x1,y1,label="Class 1",marker="o",)
plt.scatter(x2,y2,label="Class 2",marker="v",color="r")
plt.xlabel("Time Spent",fontsize=12)
plt.ylabel("Score",fontsize=12)
plt.title("Scatter Graph")
plt.legend()
plt.show()
```

```python
#Graph 4
plt.figure(figsize=(10,6))
x=np.random.normal(0,10,1000)
y=np.random.normal(0,10,1000) #we are using normal distibution of random numbers
plt.scatter(x,y)              #so that our scatter plot is distributed mostly in the middle  
plt.show()
```

```python
plt.figure(figsize=(8,6))
x=np.random.random(10)
y=np.random.random(10)
#"alpha is used for softening colors"
plt.scatter(np.random.random(10),np.random.random(10),color="red",s=50,alpha=0.6,label="one")
plt.scatter(np.random.random(10),np.random.random(10),color="b",s=100,alpha=0.8,label="two")
plt.scatter(np.random.random(10),np.random.random(10),color="g",s=150,alpha=1,label="three")
plt.scatter(np.random.random(10),np.random.random(10),color="y",s=200,alpha=0.4,label="four")
plt.legend(bbox_to_anchor=(1.0,1.0),shadow=True,fontsize='large')
plt.show()
```

# Pie Charts

```python
plt.figure(figsize=(9,9))
area=[48,30,20,15]
label=["low","medium","high","critical"]
color=["r","b","y","g"]
plt.pie(area,colors=color,labels=label,startangle=45)
plt.show()
```

```python
plt.figure(figsize=(9,9))
area=[48,30,20,15]
label=["low","medium","high","critical"]
color=["r","b","y","g"]
plt.pie(area,colors=color,labels=label,startangle=45,shadow=True,autopct="%.2f")
plt.show()
```

```python
plt.figure(figsize=(9,9))
area=[48,30,20,15]
label=["low","medium","high","critical"]
color=["r","b","y","g"]
plt.pie(area,colors=color,labels=label,startangle=45,shadow=True,autopct="%.2f",explode=[0.1,0,0.5,0])
plt.show()
```

# Donut Plot

```python
plt.figure(figsize=(9,9))
area=[48,30,20,15]
label=["low","medium","high","critical"]
color=["r","b","y","g"]
plt.pie(area,colors=color,labels=label,startangle=45)
#how to add a circle into the cente of the pie chart:
my_circle=plt.Circle((0,0),0.7,color="white")
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

```python
fig=plt.figure(figsize=(9,9))
fig.patch.set_facecolor("#DADADA") #this changes the background color of the chart
area=[48,30,20,15]
label=["low","medium","high","critical"]
color=["r","b","y","g"]
plt.pie(area,colors=color,labels=label,startangle=45)
my_circle=plt.Circle((0,0),0.7,color="white")
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

```python
fig=plt.figure(figsize=(9,9))
fig.patch.set_facecolor("#DADADA") 
area=[48,30,20,15]
label=["low","medium","high","critical"]
color=["r","b","y","g"]
plt.pie(area,colors=color,labels=label,startangle=45,explode=[0,0.2,0,0.1])
my_circle=plt.Circle((0,0),0.7,color="white")
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

```python
fig=plt.figure(figsize=(20,6))
area=[48,30,20,15]
priority=["low","medium","high","critical"]
status=["Resolved","Cancelled","Pending","Assigned"]
company=["IBM","Microsoft","Apple","Scorpion"]
colors=["r","b","y","g"]

plt.subplots=(1,3,1)
plt.pie(area,labels=priority,colors=colors,startangle=45)
my_circle=plt.Circle((0,0),0.7,color="white")
p=plt.gcf()
p.gca().add_artist(my_circle)

plt.subplots=(1,3,2)
plt.pie(area,labels=status,colors=colors,startangle=45)
my_circle=plt.Circle((0,0),0.7,color="white")
p=plt.gcf()
p.gca().add_artist(my_circle)

plt.subplots=(1,3,3)
plt.pie(area,labels=company,colors=colors,startangle=45)
my_circle=plt.Circle((0,0),0.7,color="white")
p=plt.gcf()
p.gca().add_artist(my_circle)

plt.legend()
plt.show()
```
