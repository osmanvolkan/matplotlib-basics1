# Histogram

```python
x=np.random.normal(size=2000)
plt.hist(x,bins=40,color="yellowgreen" ) #bins barların sayısıdır
plt.gca().set(title="Histogram",ylabel="Frequency") #title,x label,ylabel
plt.show()                                         #can also be inserted here
```

```python
x=np.random.rand(2000)
plt.hist(x,bins=30,color="yellowgreen" ) #bins barların sayısıdır
plt.gca().set(title="Histogram",ylabel="Frequency") #title,x label,ylabel
plt.show()                                         #can also be inserted here
```

```python
#Changing edge colors
plt.figure(figsize=(10,8))
x=np.random.normal(size=2000)
plt.hist(x,bins=40,color="orange",edgecolor="#6A9662")
plt.gca().set(title="Histogram",ylabel="Frequency") 
plt.show()                                         
```

```python
#multiple charts can be produced without using subplot method
plt.figure(figsize=(10,8))
x=np.random.normal(size=2000)
plt.hist(x,bins=40,color="orange",edgecolor="#6A9662")
plt.gca().set(title="Histogram",ylabel="Frequency") 
plt.show() 
plt.figure(figsize=(10,8))
x=np.random.normal(size=2000)
plt.hist(x,bins=20,color="blue",edgecolor="black")
plt.gca().set(title="Histogram",ylabel="Frequency") 
plt.show() 
plt.figure(figsize=(10,8))
x=np.random.normal(size=2000)
plt.hist(x,bins=10,color="pink",edgecolor="black")
plt.gca().set(title="Histogram",ylabel="Frequency") 
plt.show() 
```

```python
#Plotting multiple histograms
plt.figure(figsize=(8,11))
x=np.random.normal(-4,1,800)
y=np.random.normal(-0,1.5,800)
z=np.random.normal(3.5,1,800)
plt.hist(x,bins =30,alpha=0.6,color="yellowgreen")
plt.hist(y,bins =30,alpha=0.6,color="blue")
plt.hist(z,bins =30,alpha=0.6,color="pink")
plt.gca().set(title="Histogram",ylabel="Frequency") 
plt.show() 
```

```python
#using histogram to plot a cumilative distribution function
plt.figure(figsize=(10,8))
x=np.random.rand(2000)
plt.hist(x,bins=30,color="orange",edgecolor="black",cumulative=True)
plt.gca().set(title="Histogram",ylabel="Frequency") 
plt.show() 
```

# Area Plot

```python
x=np.arange(1,31)
y=np.random.normal(10,11,size=30)
y=np.square(y)
plt.figure(figsize=(16,6))
plt.fill_between(x,y)
plt.show()
```

```python
x=np.arange(1,31)
y=np.random.normal(10,11,size=30)
y=np.square(y)
plt.figure(figsize=(16,6))
plt.fill_between(x,y,alpha=0.2)#changed fill color transparency
plt.plot(x,y,color="green")#added edge color
plt.title("Area Chart",fontsize=16)
plt.ylabel("X",fontsize=16)
plt.ylabel("Y",fontsize=16)
plt.show()
```

```python
x=np.arange(1,6)
y1=np.array([1,5,9,13,17])
y2=np.array([2,6,10,14,16])
y3=np.array([3,7,11,15,19])
y4=np.array([4,8,12,16,20])
plt.figure(figsize=(8,6))
plt.stackplot(x,y1,y2,y3,y4,labels=["Y1","Y2","Y3","Y4"])
plt.legend(loc="upper left")
plt.show()

```

```python
x=np.arange(1,7)
y=[[1,5,9,3,17,1],[2,6,10,4,16,2],[3,7,11,5,9,1],[4,8,12,6,20,2]]
plt.figure(figsize=(10,6))
plt.stackplot(x,y,labels=["Y1","Y2","Y3","Y4"])
plt.legend(loc="upper left")
plt.show()
```
