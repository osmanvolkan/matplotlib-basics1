```python
#plotting multible sets of data
plt.figure(figsize=(10,5))
x=np.array([1,2,3,4,5,6])
y1=np.power(x,2)
y2=np.power(x,3)
plt.plot(x,y1,"b-",label="$y1=x^2$") #setting up legends
plt.plot(x,y2,"r-",label="$y2=x^3$") 
plt.legend()
plt.tight_layout()
plt.show()
```

```python
x=np.linspace(0,10,1000)
plt.figure(figsize=(10,6))
plt.plot(x,np.sin(x),label="Sin(x)") 
plt.plot(x,np.cos(x),label="Cos(x)")
plt.xlabel("X", fontsize=18)
plt.ylabel("Y", fontsize=12)
plt.title("Sin(x)",fontsize=14) #added titles and changed fontsize
plt.legend(loc="upper right")#legend will be place at upper right
plt.show()
```

```python
#changing the line style and setting up legends

plt.figure(figsize=(10,8))
x=np.array([1,2,3,4,5,6])
y1=np.power(x,2)
y2=np.power(x,3)
plt.plot(x,y1,"b-",label="$x^2$")
plt.plot(x,y2,label="$x^3$",color="red",linewidth=3.0,linestyle="--")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(loc ="upper center",fontsize="large")
plt.show()
```

```python
#line styling
x=np.linspace(0,10,2000)
plt.figure(figsize=(16,9))
plt.plot(x,np.sin(x),label="Sin(x) Dashed",linestyle="dashed")
plt.plot(x+1,np.sin(x),label="Sin(x+1) Solid",linestyle="solid")
plt.plot(x,np.cos(x),label="Cos(x) Dashed",linestyle="dashed")
plt.plot(x+1,np.cos(x),label="Cos(x+1) Dotted",linestyle="dotted")

plt.xlabel("X",fontsize=18)
plt.ylabel("Y",fontsize=18)
plt.title("Sin(x) & Cos(x)", fontsize=14)

plt.legend(loc="upper right",fontsize=14,bbox_to_anchor=(1.2,1.0))
plt.show() 
```

```python
#shading regions with will between() function

x=np.linspace(0,10,2000)
plt.figure(figsize=(10,6))
plt.plot(np.sin(x),label="Sin(x)")
plt.plot(np.cos(x),label="Cos(x)")
plt.fill_between(x,0,np.sin(x))
plt.fill_between(x,0,np.cos(x))
plt.xlabel("X",fontsize=14)
plt.ylabel("X",fontsize=14)
plt.title("Sin(x) & Cos(x)",fontsize=14)
plt.legend(loc="lower left")
plt.show()
```

```python
#changing line width
plt.figure(figsize=(10,6))
x=[1,2,3,4,5,6,7,8,9]
y=[7,8,9,10,7,11,12,13,14]
y2=[7,7,7,7,7,7,7,7,7]
plt.plot(x,y,linewidth=3,label="Line-1")
plt.plot(x,y2,linewidth=5,label="Line-2")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Line Graph")
plt.legend()
plt.show()
```

```python
#Setting the background color
plt.figure(figsize=(12,5))
x=np.array([1,2,3,4,5,6])
y1=np.power(x,2)
y2=np.power(x,3)
ax=plt.axes()
ax.set_facecolor("yellow") #setting the background colur using Hex Code
plt.plot(x,y1,"bo",x,y2,"r:") #multiple plots can be drawn using the same line
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
#Display multiple plots in one figure (1 row and two columns)
plt.figure(figsize=(14,6))
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
plt.subplot(1,2,1)#telling python wher to place graph 1
plt.plot(x,y1)
plt.subplot(1,2,2)#telling python wher to place graph 2
plt.plot(x,y2)
plt.show()
```

```python
#Display multiple plots in one figure (2 row 1 columns)
plt.figure(figsize=(14,6))
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
plt.subplot(2,1,1)#telling python wher to place graph 1
plt.plot(x,y1)
plt.subplot(2,1,2)#telling python wher to place graph 2
plt.plot(x,y2)
plt.show()
```

```python
#Displaying in multiple plots using subplots()
x=np.arange(-50,50)
y1=np.power(x,2)
y2=np.power(x,3)
y3=np.sin(x)
y4=np.cos(x)
y5=np.tan(x)
y6=np.tanh(x)
y7=np.sinh(x)
y8=np.cosh(x)
y9=np.exp(x)

fig1,ax1=plt.subplots(nrows=3,ncols=3,figsize=(20,20))
#figure with 9 places for each y is defined


ax1[0,0].plot(x,y1,"tab:blue")
ax1[0,0].set_title("Square Function") #we added set_ to these
ax1[0,0].set_xlabel("X",fontsize= 18) #because we are operating
ax1[0,0].set_ylabel("Y",fontsize= 18)   #inside a subplot
ax1[0,1].plot(x,y2,"ro")
ax1[0,1].set_title("Cubic Function")
ax1[0,1].set_xlabel("X",fontsize=18)
ax1[0,1].set_ylabel("Y",fontsize=18)
ax1[0,2].plot(x,y3,"tab:orange")
ax1[0,2].set_title("Sine Function")
ax1[0,2].set_xlabel("X",fontsize=18)
ax1[0,2].set_ylabel("Y",fontsize=18)
ax1[1,0].plot(x,y4,"g:")
ax1[1,0].set_title("Cosine Function")
ax1[1,0].set_xlabel("X",fontsize=18)
ax1[1,0].set_ylabel("Y",fontsize=18)
ax1[1,1].plot(x,y5,"b-")
ax1[1,1].set_title("Tangent Function")
ax1[1,1].set_xlabel("X",fontsize=18)
ax1[1,1].set_ylabel("Y",fontsize=18)
ax1[1,2].plot(x,y6,"g--")
ax1[1,2].set_title("Hyperbolic Tangent Function")
ax1[1,2].set_xlabel("X",fontsize=18)
ax1[1,2].set_ylabel("Y",fontsize=18)
ax1[2,0].plot(x,y7,"mv")
ax1[2,0].set_title("Hyperbolic Sine Function",fontsize=30)
ax1[2,0].set_xlabel("X",fontsize=18)
ax1[2,0].set_ylabel("Y",fontsize=18)
ax1[2,1].plot(x,y8,"y*")
ax1[2,1].set_title("Hyperbolic Cosine Function",fontsize=30)
ax1[2,1].set_xlabel("X",fontsize=30)
ax1[2,1].set_ylabel("Y",fontsize=30)
ax1[2,2].plot(x,y9,"k^")
ax1[2,2].set_title("Exp Function")
ax1[2,2].set_xlabel("X",fontsize=18)
ax1[2,2].set_ylabel("Y",fontsize=18)
plt.show()
```

# Bar Graphs

```python
id1=np.arange(1,10)
score=np.arange(20,110,10)
plt.bar(id1,score)
plt.xlabel("Student ID")
plt.ylabel("Score")
plt.show()

```

```python
id1=np.arange(1,10)
score=np.arange(20,110,10)
plt.figure(figsize=(8,5))


ax=plt.axes()
ax.set_facecolor("#ECF0F1") #setting the background colorby specifying hex code
plt.bar(id1,score,color="green")


plt.xlabel("Student ID")
plt.ylabel("Score")
plt.show()
```

```python
#plotting multiple sets of data
x1=[1,3,5,7]
x2=[2,4,6,8]
y1=[7,7,7,7]
y2=[17,18,19,40]
plt.figure(figsize=(8,6)) 
ax=plt.axes()
ax.set_facecolor("white")
plt.bar(x1,y1,label="First",color="#428300") #first set of data
plt.bar(x2,y2,label="Second",color="#94E413") #second set of data
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Bar Chart")
plt.legend()
plt.show()
```

```python
#horizontal bar chart
Age=[28,33,43,45,57]
Name=["Isaac","Asimov","John","Ronald","Raul"]
plt.barh(Name,Age,Color="yellow")
plt.show()
```

```python
#Changing Bar Width
num1=np.array([1,3,5,7,9])
num2=np.array([2,4,6,8,10])
plt.figure(figsize=(12,8))
plt.bar(num1,num1**2,width=0.2,color="#FF6F00") 
plt.bar(num2,num2**2,width=0.4,color="#FF6F00")
plt.show()
```

```python
#Displaying the values at the top of vertical bars
num1=np.array([1,3,5,7,9])
num2=np.array([2,4,6,8,10])
plt.figure(figsize=(12,8))
plt.bar(num1,num1**2,width=0.4,color="#FF6F00") 
plt.bar(num2,num2**2,width=0.4,color="#FF6F00")

for x,y in zip(num1,num1**2):
    plt.text(x,y,"%d" % y,ha="center",va="bottom")
for x,y in zip(num2,num2**2):
    plt.text(x,y,"%d" % y,ha="center",va="bottom")
             
             
plt.show()    
   
```

```python
x=np.arange(1,21)
plt.figure(figsize=(16,8)) 

y1=np.random.uniform(0.1,0.7,20)
y2=np.random.uniform(0.1,0.7,20)

plt.bar(x,+y1,facecolor="#C0CA33",edgecolor="black")
plt.bar(x,-y2,facecolor="#FF9800",edgecolor="black")

for x,y in zip(x,y1):
    plt.text(x,y,"%.2f" % y, ha="center",va="bottom",fontsize=10)

             
plt.xlim(0,21) #xlerin aralÄ±ÄŸÄ±nÄ± tekrar belirliyoruz.
plt.ylim(-1.25,1.25)    #ylerin aralÄ±ÄŸÄ±nÄ± tekrar belirliyoruz.         
             
plt.show()             
             
```
